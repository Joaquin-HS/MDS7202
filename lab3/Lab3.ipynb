**Respuesta:**

````python
def clean_data(df):
    print(f"El dataframe tiene {df.shape[0]} filas y {df.shape[1]} columnas.\n")

    # En primer lugar se eliminan registros con nombres duplicados, ya que no es posible que un alumno haya
    # dado m√°s de una vez el mismo examen. As√≠, se est√° trabajando solo con registros v√°lidos.
    print("Se eliminan registros duplicados.")
    df = df.drop_duplicates(subset=["names"], keep="first")
    print(f"El dataframe tiene {df.shape[0]} filas y {df.shape[1]} columnas.\n")

    # El segundo paso ser√≠a eliminar registros con valores nulos (de haber), ya que se est√° trabajando con
    # registros v√°lidos pero por alguna raz√≥n algunas celdas no est√°n completas.
    print("Se eliminan registros con valores nulos.")
    df.dropna(inplace=True)
    print(f"El dataframe tiene {df.shape[0]} filas y {df.shape[1]} columnas.\n")

    # Por √∫ltimo, teniendo registros v√°lidos y completos, se filtran para dejar solo aquellas filas
    # que poseen valores dentro del rango admitido de notas (entre 0 y 100).
    print("Se filtran las filas con valores dentro del rango admitido de notas (entre 0 y 100).")
    df = df[
        (df["math score"] >= 0) & (df["math score"] <= 100)
        & (df["reading score"] >= 0) & (df["reading score"] <= 100)
        & (df["writing score"] >= 0) & (df["writing score"] <= 100)
    ]
    print(f"El dataframe tiene {df.shape[0]} filas y {df.shape[1]} columnas.\n")

    return df
````

### 3.2 Obtenci√≥n de promedio de alumnos [0.3 Puntos]

Con todas las notas adjuntadas, la prestigiosa Universidad De la Cacha√±a necesita obtener el promedio general de cada uno de los alumnos. Para esto se le solicita que calcule el promedio con cada una de las asignaturas y el promedio se anexe como una nueva columna al final del `DataFrame` original. Cabe se√±alar que las notas deben ser redondeadas al segundo decimal.

**To-Do:**
- [ ] Crear funci√≥n que nos permite obtener el promedio de cada uno de los alumnos y anexar esta informaci√≥n en el `DataFrame` original.


**Respuesta:**

````python
def alumnos_promedio(dataframe_in):
    # Notas redondeadas al segundo decimal
    dataframe_in[["math score", "reading score", "writing score"]] = dataframe_in[["math score", "reading score", "writing score"]].round(2)

    # Se calcula el promedio de las tres notas, redondeado al segundo decimal
    dataframe_in["GPA"] = dataframe_in[["math score", "reading score", "writing score"]].mean(axis=1).round(2)

    return dataframe_in  # Se podr√≠a no haber retornado igual que antes. Este enfoque es distinto pero funciona igual.
````

### 3.3 Reprobados y Destacados [0.9 Puntos, 0.3 cada DataFrame]

Contentos con su desempe√±o, la Universidad les solicita que obtengan los alumnos quienes han reprobado el a√±o y los que han tenido desempe√±o de destacado. Para esto se considerar√° que un alumno ha reprobado el a√±o cuando el promedio de notas es inferior a 4.0 en la Escala Chilena, y se considerar√° destacado si posee notas superiores a 5.8 en todas las asignaturas (o sea notas superiores a 5.8 en `math score`, `reading score` y `writing score`).

Cree una funci√≥n que permita obtener tres Dataframes:
- El primero deber√° poseer solamente los alumnos que han reprobado en la prestigiosa universidad.
- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.
- El tercero ser√° un Data Frame con los alumnos destacados ordenados de mayor a menor.

**To-Do:**

- [ ] Obtener DataFrame con los alumnos que reprobaron el a√±o.
- [ ] Obtener DataFrame similar al original, pero marcados con una "R" el atributo GPA de los alumnos reprobados.
- [ ] Obtener DataFrame con los alumnos destacados ordenados de mayor a menor.

**Respuesta:**

````python
def reprobados_destacados(dataframe_in):
    # Se identifican los alumnos reprobados (promedio de notas inferior a 4.0):
    dataframe1 = dataframe_in[dataframe_in["GPA"] < 4.0]

    # Se entrega un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados:
    dataframe2 = dataframe_in.copy()
    dataframe2["GPA"] = dataframe2["GPA"].apply(lambda x: "R" if x < 4.0 else x)

    # Se identifican los alumnos destacados (notas superiores a 5.8 en todas las asignaturas) ordenados de mayor a menor:
    dataframe3 = dataframe_in[
        (dataframe_in["math score"] > 5.8)
        & (dataframe_in["reading score"] > 5.8)
        & (dataframe_in["writing score"] > 5.8)
    ].sort_values(by="GPA", ascending=False)

    return dataframe1, dataframe2, dataframe3
````

````python
dataframe1, dataframe2, dataframe3 = reprobados_destacados(df_grades)

print("Alumnos reprobados (promedio de notas inferior a 4.0):")
display(dataframe1)
print("\nAlumnos marcando con R el GPA para alumnos reprobados:")
display(dataframe2)
print("\nAlumnos destacados (notas superiores a 5.8 en todas las asignaturas) ordenados de mayor a menor:")
display(dataframe3)
````

### 4.2 An√°lisis por Nivel Educacional Etnia de los Padres [0.8 Puntos]

El rector, basado en su experiencia, cree fuertemente que el nivel educacional y la etnia de los padres influyen en las notas que obtienen sus hijos.
Como cient√≠ficos de datos, ud. y su equipo creen que deben encontrar evidencia para confirmar o refutar la hip√≥tesis del rector.

Para esto, deciden generar dos an√°lisis: una tabla de resumen por una parte y gr√°ficos de caja por otro.


**Tabla de Resumen [0.5 Puntos]**


Para generar la tabla de resumen:

- [ ] Calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average).
- [ ] Hacer una simplificaci√≥n a trav√©s de un mapeo (investigar el m√©todo `map()`) de la variable `parental level of education`  seg√∫n la siguiente conversi√≥n:

        some high school -> school
        some college -> school
        high school -> school
        bachelor's degree -> college
        associate's degree -> college
        master's degree -> postgraduate

  Los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.


- [ ] Agregar seg√∫n 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener el promedio de las notas.
- [ ] Agregar seg√∫n 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count.
- [ ] Obtener el porcentaje de alumnos con respecto al total. Los porcentajes deben ser strings que contienen la frecuencia de cada grupo con respecto al total y deben ser terminados en '%'.

Utilizar la tabla de resultados esperados como gu√≠a para desarrollar este punto.


**An√°lisis [0.3 Puntos]**

Finalmente, responda las siguientes preguntas sobre el gr√°fico de caja que viene implementado a continuaci√≥n. En este se muestra el GPA con respecto al nivel educacional, donde la variable de color es la etnicidad.

- [ ]  ¬øHay alguna diferencia entre los grupos graficados tanto para el nivel educacional de los padres como tambi√©n para la etnicidad?


**Respuesta:**

````python
def tabla_resumen(df):
    # Se genera una tabla resumen con lo solicitado

    # Se calcula el promedio de las notas de cada asignatura
    df["GPA"] = df[["math score", "reading score", "writing score"]].mean(axis=1).round(2)

    # Se mapea la variable "parental level of education":
    df["simple parental level of education"] = df["parental level of education"].map({
        "some high school": "school",
        "some college": "school",
        "high school": "school",
        "bachelor's degree": "college",
        "associate's degree": "college",
        "master's degree": "postgraduate"
    })

    # Se agrega seg√∫n "race/ethnicity" y "simple parental level of education" obteniendo el promedio de notas
    prom = df.groupby(["race/ethnicity", "simple parental level of education"])[["math score", "reading score", "writing score", "GPA"]].mean().round(2)

    # Se agrega seg√∫n "race/ethnicity" y "simple parental level of education" obteniendo el conteo de alumnos
    cont = df.groupby(["race/ethnicity", "simple parental level of education"]).size().reset_index(name='count')

    # Se unen ambos DataFrames generando la tabla resumen
    resumen = pd.merge(prom, cont, on=["race/ethnicity", "simple parental level of education"])
           
    # Se obtiene el porcentaje de alumnos con respecto al total
    total_alumnos = df.shape[0]
    resumen['percentage'] = (resumen['count'] / total_alumnos * 100).round(2).astype(str) + '%'

    return resumen
````

````python
resumen = tabla_resumen(df_grades)
resumen
````

Con respecto a la etnicidad de los padres se puede decir que los grupos B y E cuentan con mejores valores de GPA en general, mientras que los grupos A, C y D tienden a tener promedios m√°s bajos y con mayor variabilidad. ### 4.3 Joins, joins y m√°s joins [1 punto]

Mientras le notificaba por videollamada los resultados de sus descubrimientos a Don Caguayo, un exaltado practicante del area de TI entra a la reuni√≥n y les informa que ha encontrado una nueva base de datos que cuenta con las notas de dos asignaturas (en escala chilena): historia y ciencias.
Para m√°s remate, antes de huir, el practicante les cuenta que este dataframe lamentablemente contiene registros de alumnos que ya no forman parte de la universidad, por lo que no deben ser parte de este an√°lisis.


El rector (evidentemente molesto por la situaci√≥n) les ruega incluir estos datos (vaciados en el archivo other_grades.csv) al estudio original(`students_grades.csv`).


Para esto, carge el archivo `other_grades.csv` y busque la forma de unir ambos DataFrames, de tal manera que las columnas de `history score` y `science score` se anexen al final del DataFrame original. **NO LIMPIE LOS DATOS**, si no que explore los distintos tipos de merge para encontrar el mas situable para su situaci√≥n (y as√≠ evitar buscar duplicados o **alumnos que no pertenecen a la universidad**).

**To-Do**


- [ ] Cargar el `other_grades.csv`
- [ ] Unir `df_grades` con `other_grades.csv` usando outer join y explique el resultado.
- [ ] Unir `df_grades` con `other_grades.csv` usando left join y explique el resultado.
- [ ] Unir `df_grades` con `other_grades.csv` usando right join y explique el resultado.
- [ ] Unir `df_grades` con `other_grades.csv` usando inner join y explique el resultado.
- [ ] Defina cu√°l join es el que utilizar√° para generar el nuevo DataFrame.

> **Hint**: Puede explicar los resultados del merge a trav√©s de la cantidad de filas resultantes y los valores que estas contienen.

**Respuesta:**

````python
# Se cargan los archivos
students = pd.read_json('students_grades.json')
other_grades = pd.read_csv('other_grades.csv')

# Outer join (se incluyen todos los estudiantes, independiente de si est√°n en ambas tablas)
df_grades_outer = students.merge(other_grades, how='outer', left_on='names', right_on='names')

# Left join (se incluyen todos los estudiantes del DataFrame original)
df_grades_left = students.merge(other_grades, how='left', left_on='names', right_on='names')

# Right join (se incluyen todos los estudiantes del DataFrame de otras notas)
df_grades_right = students.merge(other_grades, how='right', left_on='names', right_on='names')

# Inner join (se incluyen solo los estudiantes que est√°n en ambos DataFrames)
df_grades_inner = students.merge(other_grades, how='inner', left_on='names', right_on='names')
````

````python
print("Outer join:", df_grades_outer.shape)
display(df_grades_outer)
print("Left join:", df_grades_left.shape)
display(df_grades_left)
print("Right join:", df_grades_right.shape)
display(df_grades_right)
print("Inner join:", df_grades_inner.shape)
display(df_grades_inner)
````

En primer lugar, dado que hay que considerar `students_grades.json`, se destaca que los registros no est√°n limpios ni en un formato correcto (formato chileno, con duplicados, etc). Obviando este punto, se muestra a continuaci√≥n un an√°lisis para cada tipo de join:

- Outer join: Este formato de join incluye todos los estudiantes de ambos dataframes, es decir, los que est√°n en ambos y en uno o el otro de manera individual. Esto genera un dataframe con un gran n√∫mero de registros, donde muchos de ellos contienen nulos.

- Left join: Este formato de join incluye todos los estudiantes del dataframe original, anexando las notas de `science` y `history` solo si hay coincidencias con el nombre.

- Right join: De manera similar al left join, el right join incluye todos los estudiantes del dataframe asociado a las notas de `science` y `history`, agregando la informaci√≥n de `students_grades` solo si hay coincidencias en los nombres.

- Inner join: Este tipo de join incluye solamente a los estudiantes que est√°n en ambos dataframes.

Se considera que el join m√°s adecuado ser√≠a el **inner join**, ya que permite asegurar que solo se incluyan los estudiantes que est√°n en ambos dataframes, evitando registros asociados a personas que no est√°n en la universidad (limpiando filas de ambos dataframes). Es importante destacar que esto no necesariamente generar√≠a el dataframe final totalmente limpio, pero ayuda a reducir un gran n√∫mero de problemas de limpieza que puede tener cada dataframe.
