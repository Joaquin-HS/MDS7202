> Utilice la función creada en el punto 1 para verificar sus resultados.

**Respuesta:**

````python
def clean_data(df):
    print(f"El dataframe tiene {df.shape[0]} filas y {df.shape[1]} columnas.\n")

    # En primer lugar se eliminan registros con nombres duplicados, ya que no es posible que un alumno haya
    # dado más de una vez el mismo examen. Así, se está trabajando solo con registros válidos.
    print("Se eliminan registros duplicados.")
    df = df.drop_duplicates(subset=["names"], keep="first")
    print(f"El dataframe tiene {df.shape[0]} filas y {df.shape[1]} columnas.\n")

    # El segundo paso sería eliminar registros con valores nulos (de haber), ya que se está trabajando con
    # registros válidos pero por alguna razón algunas celdas no están completas.
    print("Se eliminan registros con valores nulos.")
    df.dropna(inplace=True)
    print(f"El dataframe tiene {df.shape[0]} filas y {df.shape[1]} columnas.\n")

    # Por último, teniendo registros válidos y completos, se filtran para dejar solo aquellas filas
    # que poseen valores dentro del rango admitido de notas (entre 0 y 100).
    print("Se filtran las filas con valores dentro del rango admitido de notas (entre 0 y 100).")
    df = df[
        (df["math score"] >= 0) & (df["math score"] <= 100)
        & (df["reading score"] >= 0) & (df["reading score"] <= 100)
        & (df["writing score"] >= 0) & (df["writing score"] <= 100)
    ]
    print(f"El dataframe tiene {df.shape[0]} filas y {df.shape[1]} columnas.\n")

    return df
````

El segundo paso sería eliminar registros con valores nulos (de haber), ya que se está trabajando con registros válidos pero por alguna razón algunas celdas no están completas.

Por último, teniendo registros válidos y completos, se filtran para dejar solo aquellas filas que poseen valores dentro del rango admitido de notas (entre 0 y 100).

### 3.1 Transformación de notas a "Nota Chilena" [0.3 Puntos]

En una de las reuniones con Don Caguayo este les comenta su intención de pasar todas sus notas a la "Escala Chilena". Para esto les facilita la siguiente ecuación:

$$Nota \, Chilena= \% \, de \, logro*0.06 + 1$$

Utilizando la ecuación entregada, cree una función que le permita pasar a la escala Chilena las notas del dataframe y entregue las notas redondeadas al segundo decimal.

**To-Do**

- [ ] Tranforme los valores a la escala chilena.

**Respuesta:**

````python
def nota_chilena(dataframe_in, columna):
    dataframe_in[columna] = dataframe_in[columna].apply(lambda x: round(x*0.06 + 1, 2))  # "x" es el % de logro
    # No se retorna nada, ya que se modifica el dataframe directamente
````

### 3.2 Obtención de promedio de alumnos [0.3 Puntos]

Con todas las notas adjuntadas, la prestigiosa Universidad De la Cachaña necesita obtener el promedio general de cada uno de los alumnos. Para esto se le solicita que calcule el promedio con cada una de las asignaturas y el promedio se anexe como una nueva columna al final del `DataFrame` original. Cabe señalar que las notas deben ser redondeadas al segundo decimal.

**To-Do:**
- [ ] Crear función que nos permite obtener el promedio de cada uno de los alumnos y anexar esta información en el `DataFrame` original.


**Respuesta:**

````python
def alumnos_promedio(dataframe_in):
    # Notas redondeadas al segundo decimal
    dataframe_in[["math score", "reading score", "writing score"]] = dataframe_in[["math score", "reading score", "writing score"]].round(2)

    # Se calcula el promedio de las tres notas, redondeado al segundo decimal
    dataframe_in["GPA"] = dataframe_in[["math score", "reading score", "writing score"]].mean(axis=1).round(2)

    return dataframe_in  # Se podría no haber retornado igual que antes. Este enfoque es distinto pero funciona igual.
````

````python
df_grades = alumnos_promedio(df_grades)
df_grades.head()
````

### 3.3 Reprobados y Destacados [0.9 Puntos, 0.3 cada DataFrame]

Contentos con su desempeño, la Universidad les solicita que obtengan los alumnos quienes han reprobado el año y los que han tenido desempeño de destacado. Para esto se considerará que un alumno ha reprobado el año cuando el promedio de notas es inferior a 4.0 en la Escala Chilena, y se considerará destacado si posee notas superiores a 5.8 en todas las asignaturas (o sea notas superiores a 5.8 en `math score`, `reading score` y `writing score`).

Cree una función que permita obtener tres Dataframes:
- El primero deberá poseer solamente los alumnos que han reprobado en la prestigiosa universidad.
- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.
- El tercero será un Data Frame con los alumnos destacados ordenados de mayor a menor.

**To-Do:**

- [ ] Obtener DataFrame con los alumnos que reprobaron el año.
- [ ] Obtener DataFrame similar al original, pero marcados con una "R" el atributo GPA de los alumnos reprobados.
- [ ] Obtener DataFrame con los alumnos destacados ordenados de mayor a menor.

**Respuesta:**

````python
def reprobados_destacados(dataframe_in):
    # Se identifican los alumnos reprobados (promedio de notas inferior a 4.0):
    dataframe1 = dataframe_in[dataframe_in["GPA"] < 4.0]

    # Se entrega un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados:
    dataframe2 = dataframe_in.copy()
    dataframe2["GPA"] = dataframe2["GPA"].apply(lambda x: "R" if x < 4.0 else x)

    # Se identifican los alumnos destacados (notas superiores a 5.8 en todas las asignaturas) ordenados de mayor a menor:
    dataframe3 = dataframe_in[
        (dataframe_in["math score"] > 5.8)
        & (dataframe_in["reading score"] > 5.8)
        & (dataframe_in["writing score"] > 5.8)
    ].sort_values(by="GPA", ascending=False)

    return dataframe1, dataframe2, dataframe3
````

````python
dataframe1, dataframe2, dataframe3 = reprobados_destacados(df_grades)

print("Alumnos reprobados (promedio de notas inferior a 4.0):")
display(dataframe1)
print("\nAlumnos marcando con R el GPA para alumnos reprobados:")
display(dataframe2)
print("\nAlumnos destacados (notas superiores a 5.8 en todas las asignaturas) ordenados de mayor a menor:")
display(dataframe3)
````

Para esto, les solicita que desde el `DataFrame` que posee a todos los alumnos (`DataFrame` obtenido tras la limpieza), generen una nueva columna que clasifique en quintiles de rendimiento para sus alumnos: `['Bajísimo', 'Bajo', 'Medio', 'Alto', 'Muy alto']`. Hecho esto, grafique a través de un scatter plot el nivel educacional de los padres v/s el promedio de los alumnos, utilice los cuantiles generados anteriormente para mostrar en que cuantil se encuentra cada uno de los datos.

Con esto, es en los casos borde (promedios bajísimos o muy altos) donde el nivel de educación de los padres pasa a ser relevante sobre el promedio de notas de los estudiantes.

- [ ] Hacer una simplificación a través de un mapeo (investigar el método `map()`) de la variable `parental level of education`  según la siguiente conversión:

        some high school -> school

  Los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.


- [ ] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener el promedio de las notas.
- [ ] Agregar según 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count.
- [ ] Obtener el porcentaje de alumnos con respecto al total. Los porcentajes deben ser strings que contienen la frecuencia de cada grupo con respecto al total y deben ser terminados en '%'.

**Análisis [0.3 Puntos]**

Finalmente, responda las siguientes preguntas sobre el gráfico de caja que viene implementado a continuación. En este se muestra el GPA con respecto al nivel educacional, donde la variable de color es la etnicidad.

- [ ]  ¿Hay alguna diferencia entre los grupos graficados tanto para el nivel educacional de los padres como también para la etnicidad?


**Respuesta:**

````python
def tabla_resumen(df):
    # Se genera una tabla resumen con lo solicitado

    # Se calcula el promedio de las notas de cada asignatura
    df["GPA"] = df[["math score", "reading score", "writing score"]].mean(axis=1).round(2)

    # Se mapea la variable "parental level of education":
    df["simple parental level of education"] = df["parental level of education"].map({
        "some high school": "school",
        "some college": "school",
        "high school": "school",
        "bachelor's degree": "college",
        "associate's degree": "college",
        "master's degree": "postgraduate"
    })

    # Se agrega según "race/ethnicity" y "simple parental level of education" obteniendo el promedio de notas
    prom = df.groupby(["race/ethnicity", "simple parental level of education"])[["math score", "reading score", "writing score", "GPA"]].mean().round(2)

    # Se agrega según "race/ethnicity" y "simple parental level of education" obteniendo el conteo de alumnos
    cont = df.groupby(["race/ethnicity", "simple parental level of education"]).size().reset_index(name='count')

    # Se unen ambos DataFrames generando la tabla resumen
    resumen = pd.merge(prom, cont, on=["race/ethnicity", "simple parental level of education"])
           
    # Se obtiene el porcentaje de alumnos con respecto al total
    total_alumnos = df.shape[0]
    resumen['percentage'] = (resumen['count'] / total_alumnos * 100).round(2).astype(str) + '%'

    return resumen
````

Para más remate, antes de huir, el practicante les cuenta que este dataframe lamentablemente contiene registros de alumnos que ya no forman parte de la universidad, por lo que no deben ser parte de este análisis.


El rector (evidentemente molesto por la situación) les ruega incluir estos datos (vaciados en el archivo other_grades.csv) al estudio original(`students_grades.csv`).


Para esto, carge el archivo `other_grades.csv` y busque la forma de unir ambos DataFrames, de tal manera que las columnas de `history score` y `science score` se anexen al final del DataFrame original. **NO LIMPIE LOS DATOS**, si no que explore los distintos tipos de merge para encontrar el mas situable para su situación (y así evitar buscar duplicados o **alumnos que no pertenecen a la universidad**).

- [ ] Unir `df_grades` con `other_grades.csv` usando left join y explique el resultado.
- [ ] Unir `df_grades` con `other_grades.csv` usando right join y explique el resultado.

> **Hint**: Puede explicar los resultados del merge a través de la cantidad de filas resultantes y los valores que estas contienen.

**Respuesta:**

````python
# Se cargan los archivos
students = pd.read_json('students_grades.json')
other_grades = pd.read_csv('other_grades.csv')

# Outer join (se incluyen todos los estudiantes, independiente de si están en ambas tablas)
df_grades_outer = students.merge(other_grades, how='outer', left_on='names', right_on='names')

# Left join (se incluyen todos los estudiantes del DataFrame original)
df_grades_left = students.merge(other_grades, how='left', left_on='names', right_on='names')

# Right join (se incluyen todos los estudiantes del DataFrame de otras notas)
df_grades_right = students.merge(other_grades, how='right', left_on='names', right_on='names')

# Inner join (se incluyen solo los estudiantes que están en ambos DataFrames)
df_grades_inner = students.merge(other_grades, how='inner', left_on='names', right_on='names')
````

````python
print("Outer join:", df_grades_outer.shape)
display(df_grades_outer)
print("Left join:", df_grades_left.shape)
display(df_grades_left)
print("Right join:", df_grades_right.shape)
display(df_grades_right)
print("Inner join:", df_grades_inner.shape)
display(df_grades_inner)
````

En primer lugar, dado que hay que considerar `students_grades.json`, se destaca que los registros no están limpios ni en un formato correcto (formato chileno, con duplicados, etc). Obviando este punto, se muestra a continuación un análisis para cada tipo de join:

- Outer join: Este formato de join incluye todos los estudiantes de ambos dataframes, es decir, los que están en ambos y en uno o el otro de manera individual. Esto genera un dataframe con un gran número de registros, donde muchos de ellos contienen nulos.

- Left join: Este formato de join incluye todos los estudiantes del dataframe original, anexando las notas de `science` y `history` solo si hay coincidencias con el nombre.

- Right join: De manera similar al left join, el right join incluye todos los estudiantes del dataframe asociado a las notas de `science` y `history`, agregando la información de `students_grades` solo si hay coincidencias en los nombres.

- Inner join: Este tipo de join incluye solamente a los estudiantes que están en ambos dataframes.

Se considera que el join más adecuado sería el **inner join**, ya que permite asegurar que solo se incluyan los estudiantes que están en ambos dataframes, evitando registros asociados a personas que no están en la universidad (limpiando filas de ambos dataframes). Es importante destacar que esto no necesariamente generaría el dataframe final totalmente limpio, pero ayuda a reducir un gran número de problemas de limpieza que puede tener cada dataframe.
